```python
def fact(r):
    fact=1
    while r>1:
        fact*= r
        r=r-1
    return fact

def main():
    lambdaVal = 5
    sampleSize =10000
    distribution =[]

    for i in range(0,10): # creating an exponential distribution of numbers from 0-9
        r=i
        prob  = ( math.exp(-lambdaVal) * ( lambdaVal ** r ) )/ fact(r)
        for j in range( int(prob*sampleSize) ):
            distribution.append(r)

    
    #print(distribution)
    random.shuffle(distribution)
    plt.figure(figsize=(10,10))
    plt.subplot(211)
    plt.title("exponential distribution with lambda  ")  
    plt.xlabel("Generated random Numbers")
    plt.ylabel("Occurances")
    numberSize = 5
    numbers = []
    for i in range(numberSize): # creating the numbers 
        currentSequence = []
        for j in range(10):
            randoNo = distribution[ random.randint(0,len(distribution)-1) ]
            currentSequence.append(str(randoNo))
        numbers.append(currentSequence)
    print("Below are the generated 10 digit numbers from which a random number sequence is generated")
    for i in range(len(numbers)):
        print(numbers[i])
    probsDict = dict()
    trials = 100000
    hist=[]
    # creating a 5 digit number
    for i in range(trials): # for each trial create a 5 digit number and increase its occurance in the dictionary if it were already produced
        num=0
        for i in range(numberSize):
            randoPos = distribution [ random.randint( 0,len(distribution)-1 ) ] 
            selectedNum = int(numbers[i][randoPos])
            num+= (10**(numberSize-1-i))*selectedNum
        
       
        hist.append(num)
        if num not in probsDict:
            probsDict[num] = 1
        else:
            probsDict[num]+=1

    
    plt.hist(hist,bins=1000)

    plt.subplot(212)
    plt.title("Distribution if numbers generated were truely random")  
    plt.xlabel("Generated random Numbers")
    plt.ylabel("Occurances")

    tester=[]
    randLimit = (10**numberSize) -1
    for i in range(trials):
        tester.append( random.randint(0,randLimit))

    plt.hist(tester,bins=1000)

    plt.show()
    

main()

    












```
