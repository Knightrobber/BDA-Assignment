```python
#will take a while to run
def findVarNExp(Probs,n): # prints the variance and mean of the samples
    exp=0
    var=0
    for i in sorted(Probs.keys()):
        exp += i * Probs[i]
        var += (i**2) * Probs[i]
    var = var - exp**2
    print("Obsered Expectation = ",round(exp,6), "Variance = ", round(var,6))
    print("Theoretical Expectation = ",(1.0/2) *(0+1), " Variance = ",(1.0/ (12*n) )*((1-0)**2) )



def main():
    n=100 # can chnage the value of n here
    fxTrials=100000
    Xforfx=[]
    trials = 100000
    Probs = dict()
    hist=[]
    plt.figure(figsize=(10,10))
    #plt.subplot(211)
    #plt.title("f(X)")
    

    
    for i in range(n):
        unif = np.random.uniform(size=(1))
        Xforfx.append( round(unif,2) )

    plt.hist(Xforfx,density=True)

  
    for i in range(trials):
        X=[]
        Xn=0
        for j in range(n):
            unif = np.random.uniform(size=(1))

            X.append(round(unif[0],1))
        Xn = sum(X)/n
        hist.append(Xn)
        if Xn not in Probs:
            Probs[Xn] = 1
        else:
            Probs[Xn] += 1
    x=[]
    y=[]
    
    for i in sorted(Probs.keys()):
        x.append(i)
        Probs[i] = float(Probs[i])/trials
        y.append(Probs[i])

    plt.subplot(211)
    plt.title("f(Xn)")
    plt.xlabel("Xn values")
    plt.ylabel("Occurances")
    plt.hist(hist,bins=100,density=True)

    plt.subplot(212)
    plt.title("f(X)")  
    plt.xlabel("X values")
    plt.ylabel("Occurances")
    plt.hist(Xforfx)
    
    #plt.plot(x,y)
    findVarNExp(Probs,n)
    plt.show()



main()
```
